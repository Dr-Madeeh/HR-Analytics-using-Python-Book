## 9) العمليات الحسابية الأساسية Arithmetic Operations

جمع، طرح، ضرب، قسمة.

```python
# حساب الراتب الإجمالي
basic_salary = 5000
bonus = 1000
total_salary = basic_salary + bonus
print("الراتب الإجمالي:", total_salary)

# حساب صافي الراتب بعد خصم
deduction = 200
net_salary = total_salary - deduction
print("صافي الراتب:", net_salary)

# العلاوة السنوية 10%
annual_raise_rate = 0.10
annual_raise = basic_salary * annual_raise_rate
print("العلاوة السنوية:", annual_raise)

# متوسط الراتب الشهري للسنة
yearly_salary = total_salary * 12
monthly_average = yearly_salary / 12
print("متوسط الراتب الشهري:", monthly_average)
```

## 10) الأسس وباقي القسمة ** و %

مثال يُظهر كيف يمكن ربطه بمنطق المكافآت مع سنوات الخبرة.

```python
years = 3
base_bonus = 1000

remainder = years % 2               # باقي قسمة سنوات الخبرة على 2
bonus_value = (1.10 ** years) * base_bonus

print("سنوات الخبرة:", years)
print("الباقي عند القسمة على 2:", remainder)
print("قيمة المكافأة الأسّية:", f"{bonus_value:.2f}")
```

## 11) العمليات المنطقية Logical Operations (and / or / not)

أمثلة على أهلية الترقية والحاجة للتدريب.

```python
years_experience = 5
performance_rating = 4.2

experience_eligible = years_experience >= 3
excellent_performance = performance_rating > 4.0

promotion_eligible = experience_eligible and excellent_performance
print("مؤهل للترقية:", promotion_eligible)

needs_training = (years_experience < 2) or (performance_rating < 3.0)
print("يحتاج تدريب:", needs_training)

not_new_employee = not (years_experience < 1)
print("ليس موظفاً جديداً:", not_new_employee)
```

## 12) عمليات المقارنة Comparison Operations

مقارنة رواتب موظفين.

```python
employee_a_salary = 6000
employee_b_salary = 5500

salary_equal = employee_a_salary == employee_b_salary
print("الرواتب متساوية:", salary_equal)

salary_not_equal = employee_a_salary != employee_b_salary
print("الرواتب غير متساوية:", salary_not_equal)

a_higher = employee_a_salary > employee_b_salary
print("راتب الموظف أ أعلى:", a_higher)

a_lower_equal = employee_a_salary <= employee_b_salary
print("راتب الموظف أ أقل أو يساوي:", a_lower_equal)
```

## 13) تعبير مركب لمؤشر أداء Composite Performance Index

حساب نقاط أداء مركبة ثم تصنيف مبسط.

```python
base_performance = 3.5
attendance_score = 0.9
project_completion_rate = 0.85

total_performance_score = (base_performance * 0.5) + (attendance_score * 25) + (project_completion_rate * 25)
print("نقاط الأداء الإجمالية:", total_performance_score)

high_performer = (total_performance_score >= 40) and (attendance_score >= 0.85)
average_performer = (total_performance_score >= 30) and (total_performance_score < 40)
needs_improvement = total_performance_score < 30

print("أداء متميز:", high_performer)
print("أداء متوسط:", average_performer)
print("يحتاج تحسين:", needs_improvement)
```

## 16) مثال المكافآت وأهمية الأقواس

يوضح كيف يمكن أن يغيّر وضع الأقواس معنى الحساب.

```python
base_salary = 5000
performance_bonus_rate = 0.15
attendance_bonus = 500

# بدون أقواس
wrong_calculation = base_salary + performance_bonus_rate * attendance_bonus
print("حساب (بدون أقواس):", wrong_calculation)

# مع أقواس (هنا النتيجة قد تتطابق لأن الضرب أعلى أولوية أصلاً)
correct_calculation = base_salary + (performance_bonus_rate * attendance_bonus)
print("حساب (مع أقواس):", correct_calculation)

# حساب مختلف فعلاً بسبب تغيير موضع الأقواس
complex_bonus = (base_salary * performance_bonus_rate) + attendance_bonus
print("مكافأة مركبة:", complex_bonus)
```

## 19) if-else: تأثير نوع المدينة على الراتب المقترح

```python
city_type = "Non-Metro"
base_salary = 65000

if city_type == "Metro":
    adjusted_salary = base_salary * 1.15  # زيادة 15% للمدن الكبرى
    print(f"الراتب المعدل للمدينة الكبرى: {adjusted_salary}")
else:
    adjusted_salary = base_salary
    print(f"الراتب الأساسي للمدن الأخرى: {adjusted_salary}")
```

## 20) if-elif-else: تصنيف المرشح حسب Tier الكلية

سلسلة شروط متعددة مع مخرجات التصنيف ونسبة المكافأة.

```python
college_tier = "Tier 2"
bonus_percentage = 0

if college_tier == "Tier 1":
    bonus_percentage = 15
    classification = "مرشح من الدرجة الأولى"
elif college_tier == "Tier 2":
    bonus_percentage = 10
    classification = "مرشح من الدرجة الثانية"
elif college_tier == "Tier 3":
    bonus_percentage = 5
    classification = "مرشح من الدرجة الثالثة"
else:
    bonus_percentage = 0
    classification = "مرشح غير مصنف"

print(f"التصنيف: {classification}")
print(f"نسبة المكافأة: {bonus_percentage}%")
```

## 22) الحلقات Loops: while لمتابعة طلبات التوظيف حتى رقم محدد

## 25) for مع continue: تجاهل الموظفين الموقوفين عن العمل

تخطي هذا التكرار والانتقال للتالي.

```python
employee_statuses = ["active", "active", "suspended", "active", "suspended"]

for index, status in enumerate(employee_statuses, start=1):
    if status == "suspended":
        print(f"الموظف رقم {index} موقوف عن العمل - لا يتم احتساب راتبه")
        continue
    print(f"تم احتساب راتب الموظف رقم {index}")
```

## 26) مثال تطبيقي موسّع: المرور على صفوف DataFrame (iterrows)

ورد في الفصل استخدام iterrows للمرور على بيانات المرشحين داخل ملف CSV. هنا نقدم قالباً عملياً قابلاً للتشغيل، مع تعليق سطر القراءة الأصلي وإتاحة بيانات تجريبية.

```python
# السطر الأصلي (فعّله عند توفر الملف):
# data = pd.read_csv('Case_Study_1 Template_Employee Data.csv', skiprows=2)

# بديل تجريبي يحاكي الأعمدة المشار إليها في الشرح
data = pd.DataFrame({
    "College Tier": ["Tier 1", "Tier 2", "Tier 3"],
    "Role": ["Executive", "Associate", "Executive"],
    "City Type": ["Metro", "Non-Metro", "Metro"],
    "Previous CTC": [80000, 55000, 65000],
    "Graduation marks": [85, 66, 78],
    "Exp (Months)": [18, 30, 15],
})

for index, row in data.iterrows():
    print(f"--- Employee {index+1} ---")
    college_tier = row["College Tier"]
    role = row["Role"]
    city_type = row["City Type"]
    previous_ctc = row["Previous CTC"]
    graduation_marks = row["Graduation marks"]
    experience_months = row["Exp (Months)"]

    # منطق عرضي يجمع بين مفاهيم الفصل (يمكن تعديله وفق الدراسة)
    if graduation_marks >= 80 and experience_months >= 15 and role == "Executive":
        salary_offer = 75000
    else:
        salary_offer = 60000

    print("العرض المقترح:", salary_offer)
```
